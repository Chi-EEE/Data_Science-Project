# Counting Card Distribution

Now that we have preprocessed the data, we can now view the card distribution for season 18 of CR.

```python
%pip install pandas
%pip install matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "data/out/battle_dataset/battlesStaging_12282020_WL_tagged.csv"
```

```python
df = pd.read_csv(file_path)
print("Successfully loaded the data!")
```

```python
card_counter = 0
player_types = ["winner", "loser"]
for player_type in player_types:
    for number in range(1, 8 + 1):
            card_counter += df[f"{player_type}.card{number}.id"].value_counts()
```

```python
card_counter
```

```python
card_data = pd.read_csv("card_data.csv")
```

```python
id_to_name = dict(zip(card_data['team.card1.id'], card_data['team.card1.name']))
```

```python
card_counter.rename(id_to_name, inplace=True)
```

```python
card_counter
```

Lets generate a small pie chart for the Card Distribution

```python
plt.figure(figsize=(8, 8))
plt.pie(card_counter, labels=card_counter.index, autopct='%1.1f%%', startangle=140)
plt.title('Small Pie Chart for Card Distribution')
plt.axis('equal')
plt.show()
```

Lets generate a larger pie chart for the Card Distribution since the previous was small.

```python
plt.figure(figsize=(20, 20))
plt.pie(card_counter, labels=card_counter.index, autopct='%1.1f%%', startangle=140)
plt.title('Bigger Pie Card for the Distribution of Cards')
plt.axis('equal')
```

The card names still overlap with eachother, lets use a bar chart instead.

```python
plt.figure(figsize=(18, 6))
plt.bar(card_counter.index, card_counter.values)
plt.title('Bar Chart for the Card Distribution in S18')
plt.xlabel('Card Names')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right')
plt.show()
```
